```python
#gender distribution
```

```python
plt.figure(figsize=(5,5))
sns.countplot(data = df,x="Gender")
ax=sns.countplot(data=df,x="Gender")#assign ax
ax.bar_label(ax.containers[0])#using ax label value 
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have anaylased that:
#the number of females in the data is more #then the number of males 
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(7,6))
sns.heatmap(gb,annot=True)
plt.title("Relationship between parent's education and marks of student ")
plt.show()
```

```python
#from the above chart we have consider that parent education is impaact to students marks
```

```python
gb1= df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot=True)
plt.title("Relationship between parent's maraitial status  and marks of student ")
plt.show()
```

```python
#from the above chart parents marital status is not give any impact of student marks
```

```python
#Distribution of Ethnic Groups 
```

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()


#print(groupA["EthnicGroup"])


l=["groupA","groupB","groupC","groupD","groupE"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC ["EthnicGroup"],groupD ["EthnicGroup"],groupE ["EthnicGroup"]]

#print(mlist)

plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
